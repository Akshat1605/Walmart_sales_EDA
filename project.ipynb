```python
import pandas as pd
import pymysql
from sqlalchemy import create_engine
```

```python
df=pd.read_csv("Walmart.csv", encoding_errors="ignore")
df.head()
```

```python
#check for duplicates
df.duplicated().sum()
```

```python
#remove duplicates
df.drop_duplicates(inplace=True)
df.duplicated().sum()
```

```python
df.isnull().sum()
```

```python
df.dropna(inplace=True)
```
